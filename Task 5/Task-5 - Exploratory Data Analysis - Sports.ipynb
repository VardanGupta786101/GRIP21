<b>Graphically Representing The Number of Matches Played Based On IPL Season</b>

<b>Graphically Representing the Number of the matches Played based on venues</b>

<b>Graphically Representing The Number of Matches Won based on Batting or Bowling First </b>

```python
#Representing The Number of Matches Won based on Batting or Bowling First on pie-chart

match['win_by']=np.where(match['win_by_runs']>0,'Bat first','Bowl first')

Win=match.win_by.value_counts()
labels=np.array(Win.index)
sizes = Win.values
colors = ['#FFBF00', '#FA8072']
plt.figure(figsize = (10,8))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True,startangle=90)
plt.title('Match Result',fontsize=20)
plt.axis('equal')
plt.show()
```

<b>Graphically Comparing the Numbers of matches won by batting or bowling first based on respective Sessions.</b>

```python
#Graphically Comparing the Numbers of matches won by batting or bowling first based on respective Sessions Using Stacked count-plot.

plt.figure(figsize = (18,10))
sns.countplot(x='season',hue='win_by',data=match,palette='hsv')
plt.title("Numbers of matches won by batting and bowling first ",fontsize=20)
plt.xlabel("Season",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.grid(axis='y')
plt.show()
```
