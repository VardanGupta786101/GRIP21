<h3>Importing All The Required Libraries</h3>

```python
#Importing The Required Libraries

import cv2
import matplotlib.pyplot as plt
import easyocr
from pylab import rcParams
```

```python
from IPython.display import Image
rcParams['figure.figsize'] = 10,20
```

<h3>Loading/Downloading English Detection Model</h3>

```python
English_Reader = easyocr.Reader(['en'])
```

<h3>Demo Image for Testing The Code</h3>

```python
Image('Sample.png')
```

<h3>Code For Character object Detection</h3>

<b>Reading The Text Inside Sample.png</b>

```python
Text_output = English_Reader.readtext('Sample.png')
```

<b>Print the Detected Information from 'Sample.png'</b>

```python
print(Text_output)
```

<b>This list shows the co-ordinates of bounding box , Text and Confidence level</b>

<b>Extract the Fetched Text Data  & Bounding it with box coordinates</b>

```python
cord1 = Text_output[0][0]
print(Text_output[0][1])
```

<b>Calculating Minimum coordinates</b>

```python
x_min1,y_min1 = [min(idx) for idx in zip(*cord1)]
```

<b>Calculating Maximum coordinates</b>

```python
x_max1,y_max1 = [max(idx) for idx in zip(*cord1)]
```

<h3>Visualization of bounding box co-ordinates of particular text</h3>

```python
image = cv2.imread('Sample.png')
cv2.rectangle(image,(int(x_min1),int(y_min1)),(int(x_max1),int(y_max1)),(0,255,0),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

<b>Extract the Fetched Text Data  & Bounding it with box coordinates</b>

```python

cord2 = Text_output[1][0]
print(Text_output[1][1])
```

<b>Calculating Minimum and Maximum coordinates</b>

```python
x_min2,y_min2 = [min(idx) for idx in zip(*cord2)]
x_max2,y_max2 = [max(idx) for idx in zip(*cord2)]
```

```python

image = cv2.imread('Sample.png')
cv2.rectangle(image,(int(x_min2),int(y_min2)),(int(x_max2),int(y_max2)),(0,255,0),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
