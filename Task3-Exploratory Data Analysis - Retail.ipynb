<b>Exploratory Data Analysis</b>

<b>Adding new column ProfitSalesRation</b>

```python
#Adding new columngraph named as ProfitSalesRatio

subcat_sales_profit = df.groupby(['Sub-Category'],as_index=False)[['Quantity','Sales','Discount','Profit']].sum()

#Generating Values for ProfitSalesRation
from decimal import *
def p2s(x):
    getcontext().prec=2
    val = Decimal(x['Profit'])/Decimal(x['Sales'])
    return val


#Adding The Values inside Dataset
ps=subcat_sales_profit.apply(p2s,axis=1)
ps=pd.DataFrame(data=ps,columns=['ProfitSalesRatio'])
subcat_sales_profit=pd.concat([subcat_sales_profit,ps],axis=1)

#printing subcat_sales_profitt
print (subcat_sales_profit)
```

<h3>Ploting Profit to Sales Ration In Bar Graph</h3>

```python
#coding for Bar Graph

fig=plt.figure(figsize=(20,6))
plt.bar(subcat_sales_profit['Sub-Category'],subcat_sales_profit['ProfitSalesRatio'])
plt.title("Profit to Sales Ratio",fontsize=27)
plt.show()
```

<h3>ProfitSales Ratio Shows Amount Of Profit,Loss and Total Profit Per Value of Sale of Good</h3>

```python
#Graphically Representing The Amount Of Profit,Loss and Total Profit Per Value of Sale of Good

fig2,axes= plt.subplots(3,1,figsize=(10,10),squeeze=False)
X=df['Profit']/df['Quantity']>0
X=df.loc[X,'Region']
X=X.value_counts()
Y=df['Profit']/df['Quantity']<=0
Y=df.loc[Y,'Region']
Y=Y.value_counts()
fig2.suptitle("Analysis of Profits and Loss by Region ")
sns.barplot(x= X.index,y=X,ax=axes[0][0])
sns.barplot(ax=axes[1][0],x=Y.index,y=Y)

sns.scatterplot(ax=axes[2][0],x=df['Sales']/df['Quantity'],y=(df['Profit']/df['Quantity']), data=df, hue='Region')
axes[0][0].set(xlabel="Region", ylabel="Profit Count")
axes[1][0].set(xlabel="Region", ylabel="Loss Count")
axes[2][0].set(xlabel="Sales", ylabel="Profit")
plt.legend(loc="upper right",ncol=5)
plt.tight_layout()
plt.show()
```

<h4>We can see from above graph region with most profit and most lost as well overall profit count is much higher than loss count</h4>

<h3>Calculating and Graphically Representing Variation of Price Of Different Commodity</h3>

```python
#Calculating Variation of Price Of Different CommodityÂ¶

df3=df.groupby(['State','Category','Sub-Category'],as_index=False)[['Sales','Quantity','Discount','Profit']].sum()
df3=df3.sort_values(['State','Sub-Category'])
df3=df3.reset_index(drop=True)

def calprice(y):
    getcontext().prec=4
    val =(Decimal(y['Sales']) - Decimal(y['Profit']))/ Decimal(y['Quantity'])
    return val
price=df3.apply(calprice,axis=1)
price=pd.DataFrame(data=price,columns=['Price'])
df3=pd.concat([df3,price],axis=1)
df3.head(10000)
```

```python
#Graphically Representing Variation of Price Of Different CommodityÂ¶

fig7= plt.figure(figsize=(20,10))
plt.title("Variation of Price Of Different Commodity",fontsize=27)
plt.ylabel("Price",fontsize=30)
plt.xlabel("Subcategory",fontsize=30)
sns.scatterplot(x='Sub-Category',y='Price', data=df3, hue='Sub-Category')
plt.legend(loc="upper right",ncol=5 , )
plt.show()
```

<h4>From above graph we can see that their is huge discrepency in price of same commodity across different states<br>This may be due to logistic charges assuming that price and profit margin is kept same across all the states</h4>

<h3> Heatmap Representing the quantity of product ordered by each state</h3>

```python
#Calculating The Quantity of Product Ordered by each State.

states = df3['State'].unique()
subcat=df3['Sub-Category'].unique()

rows=[]
for st in states:
    idex = df3['State']==st
    idex=np.where(idex)[0]
    for sub in subcat:
        idex2=df3['Sub-Category']==sub
        idex2 = np.where(idex2)[0]
        s=np.intersect1d(idex,idex2)
        if(s.size>0):
            rows=np.append(rows,df3['Quantity'].values[s])
        else:
            rows=np.append(rows,[0])

rows = rows.reshape(49,17)

df4=pd.DataFrame(rows)
df4.columns=subcat   
df4.index=states
```

```python
#Graphically Representing The Quantity of Product Ordered by each State on Heatmap

sns.set()
fig8 =plt.figure(figsize=(20,40))
sns.heatmap(df4,annot=True,fmt="g",cmap="viridis")
plt.show()
```

<h4>Above Heatmap shows the quantity of product ordered by each state , this can help us to recommend the products in demand in each state and introduce our new products accordingly.</h4>
